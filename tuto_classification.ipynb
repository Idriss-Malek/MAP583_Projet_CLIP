# Tutorial: Classifying images

**Importing libraries**

```python
from PIL import Image
from Classifier import Classifier
```

**Creating a classifier with the dog and cat labels and opening images**

```python
classifier=Classifier(['cat','dog'])
classifier.load_classifier('dog_and_cat.txt')
```

```python
dog=Image.open("dog.png")
cat=Image.open("cat.jpg")
```

**Trying to classify an image of a dog and a cat**

```python
classifier.set_image(dog)
```

```python
classifier.classify()
```

```python
classifier.set_image(cat)
```

```python
classifier.classify()
```

**Explaining feature**

```python
classifier.explain('cat')
```
